```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
star_wars = pd.read_csv('star_wars.csv',encoding='ISO-8859-1')
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

# Removing any null RespondentID values

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

```python
star_wars.head(10)
```

# Converting Yes/No responses to Boolean

```python
yes_no = {'Yes': True, 'No': False}
for x in star_wars.columns[1:3]:
    star_wars[x] = star_wars[x].map(yes_no)
```

```python
star_wars.head(5)
```

```python
yes_no = {"Star Wars: Episode I  The Phantom Menace": True, np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True}
for x in star_wars.columns[3:9]:
    star_wars[x] = star_wars[x].map(yes_no)
```

```python
star_wars.head(4)
```

```python
star_wars = star_wars.rename(columns={
    'Which of the following Star Wars films have you seen? Please select all that apply.':'episode 1',
    'Unnamed: 4':'episode 2','Unnamed: 5':'episode 3',
    'Unnamed: 6':'episode 4','Unnamed: 7': 'episode 5',
    'Unnamed: 8':'episode 6'
})
```

```python
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
    'episode_1': 'episode 1'
})
```

```python
star_wars.head()
```

```python

for x in star_wars.columns[9:15]:
    star_wars[x] = pd.to_numeric(star_wars[x])
```

```python
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking 1',
    'Unnamed: 10' : 'ranking 2',
    'Unnamed: 11' : 'ranking 3',
    'Unnamed: 12' : 'ranking 4',
    'Unnamed: 13' : 'ranking 5',
    'Unnamed: 14' : 'ranking 6'
})
```

```python
star_wars_ranking_mean = star_wars[star_wars.columns[9:15]].mean()
```

```python
star_wars_ranking_mean
```

```python
%matplotlib inline
```

```python
fig,ax = plt.subplots()
ax.bar([1,2,3,4,5,6],star_wars_ranking_mean,width=0.9)
ax.set_xticklabels(star_wars.columns[9:15],rotation=90)
```

As Star wars released the first three episodes later, I suspect that the newer ones in general are rated lower than the old ones

```python
star_wars_sum = star_wars[star_wars.columns[3:9]].sum()
```

```python
fig,ax = plt.subplots()
ax.bar(range(6),star_wars_sum,width=0.9)
```

It looks like the old series have been seen more. That could be even the generation difference

# Splitting by gender and exploring further

```python
males = star_wars[star_wars['Gender']=='Male']
females = star_wars[star_wars['Gender']=='Female']
```

# Exploring the most viewed and the highest ranked for each gender

```python
rankings_m = males[males.columns[9:15]].mean()
rankings_f = females[females.columns[9:15]].mean()
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.bar(range(6),rankings_m,width=0.9)
ax2.bar(range(6),rankings_f,width=0.9)
ax1.set_xlabel('Males')
ax2.set_xlabel('Females')
```

Looks like the males in general rank the older episodes higher, compared to the females
